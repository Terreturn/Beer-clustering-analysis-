## 2. Contents of the dataset

## Beer Profile and Ratings Dataset Documentation

This dataset integrates basic beer metadata, sensory profile features derived from review texts, and aggregated user ratings.

---

## Main Dataset: `beer_profile_and_ratings.csv` (`df_beer`)

The dataset consists of 25 columns, categorized into three primary groups:

### 1. Basic Information
| No. | Column Name | Description |
| :--- | :--- | :--- |
| 1 | **Name** | Beer name (Label) |
| 2 | **Style** | Beer Style |
| 3 | **Brewery** | Brewery name |
| 4 | **Beer Name (Full)** | Complete beer name (Brewery + Brew Name) — **Unique Identifier** |
| 5 | **Description** | Notes on the beer (if available) |
| 6 | **ABV** | Alcohol content by volume (%) |
| 7 | **Min IBU** | Minimum IBU value for the specific beer style |
| 8 | **Max IBU** | Maximum IBU value for the specific beer style |

### 2. Tasting Profile Features
*Note: These features represent word counts extracted from up to 25 reviews per beer. They quantify the presence of specific descriptors mentioned by reviewers.*

* **Mouthfeel:**
    * `Astringency`, `Body`, `Alcohol`
* **Taste:**
    * `Bitter`, `Sweet`, `Sour`, `Salty`
* **Flavor and Aroma:**
    * `Fruits`, `Hoppy`, `Spices`, `Malty`

### 3. Review and Rating Metrics (higher values indicate more positive reviews)
| No. | Column Name | Description |
| :--- | :--- | :--- |
| 20 | **review_aroma** | Average score for aroma |
| 21 | **review_appearance** | Average score for appearance |
| 22 | **review_palate** | Average score for palate |
| 23 | **review_taste** | Average score for taste |
| 24 | **review_overall** | Average overall rating score |
| 25 | **number_of_reviews** | Total count of consumer/user reviews |
